```python
#Q5 replace value in dataframe
df=dataSet
df.loc[df['Species'] == 'Iris-setosa', 'Species'] = 0
df.loc[df['Species'] == 'Iris-versicolor', 'Species'] = 1
df.loc[df['Species'] == 'Iris-virginica', 'Species'] = 2
```

```python
#Q6
df.head(10)
```

```python
#Q7
from sklearn.model_selection import train_test_split
data_train, data_test=train_test_split(df, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
#Q8
data_train.head(10)
```

```python
data_test.head(10)
```

```python
#Q9
from sklearn.neural_network import MLPClassifier
# lbfgs, epsilon=0.07 
clf = MLPClassifier(solver='lbfgs', max_iter=150, random_state=0, epsilon=0.07 )
```

```python
#Q10
X=data_train.iloc[:,1:5].values
Y=data_train['Species'].values.astype('int')
clf.fit(X,Y)
```

```python
#q11
from sklearn.metrics import accuracy_score
predictions = clf.predict(data_test.iloc[:,1:5].values)
print("Taux de prcision ",accuracy_score(data_test['Species'].values.astype('int'),predictions))
```

```python
#q12
from sklearn.metrics import  plot_confusion_matrix
plot_confusion_matrix(clf, data_test.iloc[:,1:5].values,data_test['Species'].values.astype('int'))
```
