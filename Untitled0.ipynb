```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/tp/market_basket.txt', sep="\t", names=['ID','Product'])
df.head(10)
```

```python
print(df.shape)
```

```python
VAR = pd.crosstab(df.ID,df.Product)
print(VAR.transpose())#transposé du res pr un affichage 5ir w mnathem    
```

```python
VAR.iloc[:30,:3] # illoc recupere les 30 1er lignes et les 3 premier colones
```

```python
from mlxtend.frequent_patterns import apriori 
#itemsets frequents
freq_itemsets = apriori(VAR,min_support=0.025,max_len=4,use_colnames=True)
```

```python
freq_itemsets.head(15)
```

```python
def is_inclus(x,items):
  return items.issubset(x)

```

```python
import numpy
id = numpy.where(freq_itemsets.itemsets.apply(is_inclus,items={'Aspirin'}))
print(id)
```

```python
id = numpy.where(freq_itemsets.itemsets.apply(is_inclus,items={'Aspirin','Eggs'}))
print(id)
```

```python
print(freq_itemsets[freq_itemsets['itemsets'].ge({'Aspirin','Eggs'})])
```

```python
from mlxtend.frequent_patterns import association_rules 
regles = association_rules(freq_itemsets,metric="confidence",min_threshold=0.75)
print(regles.iloc[:5,:])
```

```python
print(regles[regles['lift'].ge(7.0)])
```

```python
print(regles[regles['consequents'].eq({'2pct_Milk'})])
```
